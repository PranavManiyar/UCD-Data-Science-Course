```python
fig,ax=plt.subplots()
ax.plot (a,b)
ax.plot (b,a)
#ax.bar (a,b)
ax.scatter(b,a)
plt.show()
```

```python
fig,ax=plt.subplots()
ax.plot (a,b,marker="o")
ax.plot (b,a,marker="X",color="r")
#ax.bar (a,b)
#ax.scatter(b,a)
plt.show()
```
